##### Task #6 :Decision Tree classifier 

Problem Statement : Create the Decision Tree classifier and visualize it graphically.

#### Step 1 : Importing libraries 

```python
import sklearn.datasets as datasets
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

#### Step 2 : Loading dataset

```python
iris = datasets.load_iris()
# Loading x and y :
x=pd.DataFrame(iris.data[:,:], columns=iris.feature_names[:])
y=pd.DataFrame(iris.target, columns=["Species"])
```

```python
y.sample(10)
```

```python
x.sample(10)
```

#### Step 3 : Creating Decision tree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=6) 
decision_tree.fit(x, y)
```

#### Step 4 : Visualizing the Decision Tree 

For Visualizing decision tree, we require libraries :

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(40,50))
viz = tree.plot_tree(decision_tree, 
                   feature_names=x.columns,  
                   class_names=iris.target_names,
                   filled=True)
```
